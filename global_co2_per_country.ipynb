```python
import pandas as pd
import plotly.express as px # type: ignore
import plotly.graph_objects as go # type: ignore

# Läs in datan
df = pd.read_csv("cleaned_co2_data.csv")

# Lägg till kolumn med CO2 i miljoner ton
df["co2_milj_ton"] = df["value"] / 1000

# Skapa animerad karta där du kan välja år
fig = px.choropleth(
    df,
    locations="country_code",        # ISO 3-koder (t.ex. SWE, USA)
    color="co2_milj_ton",            # Färg baserat på utsläpp
    hover_name="country_name",       # Visas vid hover
    animation_frame="Year",          # Slider för att välja år
    color_continuous_scale="YlOrRd", # Färgskala
    labels={"co2_milj_ton": "CO₂ (miljoner ton)"},
    title="Globala CO₂-utsläpp per land (med årtalsreglage)",
    projection="natural earth"
)

# Justera storlek och spara
fig.update_layout(width=1000, height=600)
fig.write_html("co2_animation_map.html")  # Spara interaktiv HTML-fil
fig.show()
```
