```python
import pandas as pd
import matplotlib.pyplot as plt

# Ls in CSV-filen
df = pd.read_csv("co2_naturkatastrofer_2017_2023.csv")

# Skapa figuren och tv y-axlar
fig, ax1 = plt.subplots(figsize=(10, 5))

# Vnster y-axel: klimatkatastrofer
ax1.set_xlabel("r")
ax1.set_ylabel("Antal klimatkatastrofer", color="green")
ax1.plot(df["Year"], df["Disasters_Count"], marker="o", color="green", label="Naturkatastrofer")
ax1.tick_params(axis="y", labelcolor="green")
ax1.set_ylim(df["Disasters_Count"].min() - 10, df["Disasters_Count"].max() + 5)

# Hger y-axel: CO2-utslpp
ax2 = ax1.twinx()
ax2.set_ylabel("CO2-utslpp (miljoner ton)", color="blue")
ax2.plot(df["Year"], df["CO2_milj_ton"], marker="x", color="blue", label="CO2-utslpp")
ax2.tick_params(axis="y", labelcolor="blue")
ax2.set_ylim(df["CO2_milj_ton"].min() - 200, df["CO2_milj_ton"].max() + 100)

# Markera pandemin visuellt
ax1.axvspan(2020, 2021, color="grey", alpha=0.2, label="Covid-19")

# Titel och layout
plt.title("Klimatkatastrofer och CO2-utslpp (20172023)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
